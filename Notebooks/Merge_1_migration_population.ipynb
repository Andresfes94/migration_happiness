```python
columuns_to_use=list(migration_table.columns)
```

```python
population_to_use=population_table[columuns_to_use].copy()
```

```python
population_to_use
```

```python
clean_cols=['Indicator Name', 'Indicator Code',
       '1962', '1967', '1972', '1977', '1982', '1987', '1992', '1997', '2002',
       '2007', '2012', '2017']
```

```python
population_to_use=population_to_use[clean_cols].copy()
```

```python
population_to_use
```

```python
population_to_use.to_csv('population_to_combine.csv')
```

```python
combined_data = pd.concat([migration_table,population_to_use],
                          axis=1).reindex(migration_table.index)
#df4.reindex(df1.index)
```
